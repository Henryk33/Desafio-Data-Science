```python
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

ingresos_tienda1 = tienda1['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()

print(f"Ingresos de la tienda 1: {ingresos_tienda1}")
print(f"Ingresos de la tienda 2: {ingresos_tienda2}")
print(f"Ingresos de la tienda 3: {ingresos_tienda3}")
print(f"Ingresos de la tienda 4: {ingresos_tienda4}")
```

```python
ingresos = {'Tienda 1': ingresos_tienda1, 'Tienda 2': ingresos_tienda2, 'Tienda 3': ingresos_tienda3, 'Tienda 4': ingresos_tienda4}

ingresos_df = pd.DataFrame(list(ingresos.items()), columns=['Tienda', 'Ingresos'])

print(ingresos_df)
```

### Mayor y menor ingreso en las 4 tiendas

```python
tienda_mayor_ingreso = max(ingresos, key=ingresos.get)
print(f"La tienda con mayor ingreso es: {tienda_mayor_ingreso}")
tienda_menor_ingreso = min(ingresos, key=ingresos.get)
print(f"La tienda con menor ingreso es: {tienda_menor_ingreso}")
```

### GRAFICA COMPARATIVA:

```python
import matplotlib.pyplot as plt
from IPython.display import display

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]

colores = ['beige', 'lavender', 'lightblue', 'mistyrose']

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(tiendas, ingresos, color=colores)

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:,.0f}', ha='center', va='bottom', fontsize=12)


ax.set_ylim(0, max(ingresos) * 1.1)
ax.set_title('Comparación de ingresos totales por tiendas', fontsize=16)
ax.set_xlabel('Tiendas', fontsize=14)
ax.set_ylabel('Ingresos totales', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

#Conclusión sobre los ingresos totales por tiendas:
Gracias a los datos de las ventas de cada tienda podernos darnos cuenta que la tienda con menor ingreso ha sido la "Tienda 4" cuya comparativa nos relata que tiene una diferencia de casi $60,000,000 menos que la "Tienda 3" siendo esta la penultima en ingresos.

# 3. Calificación promedio de la tienda


```python
calificacion_tienda1 = tienda1['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio de la tienda 1: {calificacion_tienda1}")
print(f"Calificación promedio de la tienda 2: {calificacion_tienda2}")
print(f"Calificación promedio de la tienda 3: {calificacion_tienda3}")
print(f"Calificación promedio de la tienda 4: {calificacion_tienda4}")
```

```python
import numpy as np

calificacion_promedio_tienda1 = tienda1['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificacion_promedio = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

num_bars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_bars, endpoint=False).tolist()

calificacion_promedio += calificacion_promedio[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.plot(angles, calificacion_promedio, linewidth=1, linestyle='solid')

ax.fill(angles, calificacion_promedio, 'b', alpha=0.1)

ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_ylim(0, 5)

ax.set_title('Calificación promedio por tienda', va='bottom', pad=40)

plt.show()
```

```python
plt.figure(figsize=(8, 8))
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

calificacion_promedio = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

colores = ['beige', 'lavender', 'lightblue', 'mistyrose']

bars = plt.bar(labels, calificacion_promedio, color=colores)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

#Conclusión en la categoría de "Calificación por los usuarios" de cada tienda.
Gracias a la opinion general de los usuarios nos damos cuenta que la tienda 3, a pesar de no tener el mayor número de ingresos, es la mejor calificada por los usuarios.

Además que la tienda 1, a pesar de ser la tienda más rentable, cuenta con la peor calificación entre las 4.


# 4. Productos más y menos vendidos

```python
tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

productos_vendidos = tiendas.groupby('Producto')['Cantidad de cuotas'].sum().reset_index()

productos_vendidos_ordenados = productos_vendidos.sort_values(by='Cantidad de cuotas', ascending=False)

productos_vendidos_ordenados.head(), productos_vendidos_ordenados.tail()
```

```python
ventas_producto = tiendas['Producto'].value_counts()
ventas_producto.columns = ventas_producto.sort_values()

plt.figure(figsize=(10, 6))
colors = plt.cm.viridis(ventas_producto / max(ventas_producto))
plt.bar(ventas_producto.index, ventas_producto.values, color=colors)

plt.title('Productos más y menos vendidos.', fontsize=16)
plt.xlabel('producto', fontsize=12)
plt.ylabel('Cantidad de ventas', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda1 = tienda1['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
envios = [envio_promedio_tienda1, envio_promedio_tienda2, envio_promedio_tienda3, envio_promedio_tienda4]

sorted_envios = sorted(zip(envios, tiendas))
envios_sorted, tiendas_sorted = zip(*sorted_envios)

plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas_sorted, envios_sorted, color=plt.cm.viridis(range(len(tiendas_sorted))))

plt.title('Envío promedio por tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tienda', fontsize=12, fontweight='bold')
plt.ylabel('Envío promedio', fontsize=12, fontweight='bold')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
costo_envio = [10, 12, 8, 15]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, costo_envio, color=plt.cm.viridis(range(len(tiendas))))

plt.title('Costo de Envío Promedio por Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tienda', fontsize=12, fontweight='bold')
plt.ylabel('Costo de Envío Promedio (en USD)', fontsize=12, fontweight='bold')

for i, v in enumerate(costo_envio):
    plt.text(i, v, str(v), ha='center', va='bottom', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

Tienda 2: 12$

Tienda 4: 15$
El costo de envío menor sería el de la tienda 3 y el mayor el de la tienda 4.
